```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Travel.csv', delimiter="|")
df.head()
```

```python
df['MonthlyIncome'].describe()
```

segmentasi 

- Low Income: 1,000 - 34,225
- Middle Income: 34,226 - 67,451
- High Income: 67,452 - 98,678

```python
def segmentasi_income(x):
    income = x['MonthlyIncome']
    if 1000 <= income <= 34225:
        segmentasi = 'Low Income'
    elif 34226 <= income <= 67451:
        segmentasi = 'Middle Income'
    else:
        segmentasi = 'High Income'
    return segmentasi
```

```python
df['segmentasi_income'] = df.apply(lambda x: segmentasi_income(x), axis=1)
df.head()
```

```python
segment_counts_1 = df[df['ProdTaken'] == 1]['segmentasi_income'].value_counts()
colors = ['#3498db', '#5DADE2', '#85C1E9']
fig, ax = plt.subplots(figsize=(7, 8))

ax.bar(segment_counts_1.index, segment_counts_1.values, color=colors)

ax.set_xlabel('Income Segment')
ax.set_ylabel('Count')
ax.set_title('MonthlyIncome Segmentation for ProdTaken=1')

plt.show()
```

```python
low_income_counts_1 = df[(df['ProdTaken'] == 1) & (df['segmentasi_income'] == 'Low Income')].shape[0]
low_income_counts_0 = df[(df['ProdTaken'] == 0) & (df['segmentasi_income'] == 'Low Income')].shape[0]

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))

labels = ['ProdTaken=1', 'ProdTaken=0']
sizes = [low_income_counts_1, low_income_counts_0]
colors = ['#3498db', '#e74c3c']

ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.set_title('ProdTaken Distribution for Low Income Segment')

plt.show()
```

```python
low_income_counts_1 = df[(df['ProdTaken'] == 1) & (df['segmentasi_income'] == 'Middle Income')].shape[0]
low_income_counts_0 = df[(df['ProdTaken'] == 0) & (df['segmentasi_income'] == 'Middle Income')].shape[0]

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))

labels = ['ProdTaken=1', 'ProdTaken=0']
sizes = [low_income_counts_1, low_income_counts_0]
colors = ['#3498db', '#e74c3c']

ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.set_title('ProdTaken Distribution for Middle Income Segment')

plt.show()
```

```python
low_income_counts_1 = df[(df['ProdTaken'] == 1) & (df['segmentasi_income'] == 'High Income')].shape[0]
low_income_counts_0 = df[(df['ProdTaken'] == 0) & (df['segmentasi_income'] == 'High Income')].shape[0]

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))

labels = ['ProdTaken=1', 'ProdTaken=0']
sizes = [low_income_counts_1, low_income_counts_0]
colors = ['#3498db', '#e74c3c']

ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.set_title('ProdTaken Distribution for High Income Segment')

plt.show()
```

```python
# Assuming 'segmentasi_income' is a categorical variable in your DataFrame
segment_counts_1 = df[df['ProdTaken'] == 1]['segmentasi_income'].value_counts()
segment_counts_0 = df[df['ProdTaken'] == 0]['segmentasi_income'].value_counts()

# Extract unique income segments
income_segments = segment_counts_1.index.union(segment_counts_0.index)

# Set up positions for bars
bar_positions = np.arange(len(income_segments))
bar_width = 0.35

# Plot the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(bar_positions - bar_width/2, segment_counts_1.reindex(income_segments, fill_value=0), bar_width, label='ProdTaken=1', color='#3498db')
ax.bar(bar_positions + bar_width/2, segment_counts_0.reindex(income_segments, fill_value=0), bar_width, label='ProdTaken=0', color='#e74c3c')

ax.set_xticks(bar_positions)
ax.set_xticklabels(income_segments)
ax.set_xlabel('Income Segment')
ax.set_ylabel('Count')
ax.set_title('MonthlyIncome Segmentation for ProdTaken=0 and ProdTaken=1')
ax.legend()

plt.show()
```

Insight

Pada grafik diatas terlihat bahwa kategori 'Low Income' menonjol dengan jumlah tertinggi dalam kategori 'ProdTaken=1', mencapai total 856 pelanggan. Jumlah persentase customer yang melakukan pembelian produk ('ProdTaken=1') dibanding customer yang ditelfon juga terbanyak ada pada kategori 'Low Income', mencapai 19.6%.
